### creating X and Y Labels for features and tagret

```python
#creating X labes(all columns except target),whcih is features matrix
X = cancer_df.drop("Target", axis =1)

# creating Y labels that will just contain Target
Y = cancer_df["Target"]
```

```python
len(X), len(Y)
```

### Spilting Data into Traning data and testing data


```python
#Now we've split our data into X and y, we'll use Scikit-Learn to split it into training and test sets.
np.random.seed(42)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size= 0.3) # spliting data into 70% training 30%testing
```

```python
# checking all the shape of our split data
X_train.shape, X_test.shape,Y_train.shape, Y_test.shape  
```

### Preparing different models 

```python
from sklearn.ensemble import RandomForestClassifier # importing Ranodom_Forest
from sklearn.svm import LinearSVC  #importing the SVC
from sklearn.neighbors import NearestNeighbors # importing the KNN
from sklearn import svm # importing the SVM model
from sklearn.linear_model import LogisticRegression # Importing the Logistic_Regression model
from sklearn.naive_bayes import GaussianNB  # Importing the GaussainNB
```

### Creating a dictionary called models which contains all of the classification models

```python
# we create this alll the models into dic so that we can pass all the models in one time
models_dict= {"LinearSVC" :LinearSVC(),
         "SVM":svm.SVC(),
         "LogisticRegression": LogisticRegression(),
         "RandomForestClassifier": RandomForestClassifier(n_estimators=100),
         "GaussianNB": GaussianNB()}

```

### Looping through all the models_dict and use fit( ) and score() on all the model


```python
from sklearn.model_selection import cross_val_score
#Creating an empty dictionary called Result
result_dict = {}

np.random.seed(42)
# we will use cross-validation check the accuracy of each model
for model_name,i in models_dict.items():
    i.fit(X_train,Y_train)
    result_dict[model_name] =np.mean(cross_val_score(i,X,Y,cv=5, scoring = "accuracy"))
```

### This will show all the Accuracy of all Models

```python
result_dict
```

```python
#from the above Result we can say that RandomForestClassifer has the best Score Accuracy
```

### HyperParamter Tunning (RandomForestClassifer)

```python
#Now we found that RandomForestClassifier is best , we will try to Tunning
## we will choose RandomSearchCV 
```

```python
from sklearn.model_selection import RandomizedSearchCV
# we will made the grid with all the hyperparamter we need to change in the  RandomForestClassifier

np.random.seed(42)

# Making the grid of all the paramter we need to changes
grid = {"n_estimators": [10,100,200,500,1000,1500],
        "max_depth": [None,5,10,20,30],
        "min_samples_split": [2,4,6],
        "max_depth":[2,4,6,8]}


#Instantiate RandomForestClassifier 
rfc_clf =RandomForestClassifier(n_jobs =1) 

#Set up RandomSearchCV 
rs_clf = RandomizedSearchCV(estimator = rfc_clf,
                           param_distributions = grid,
                           n_iter =10, # Number of time the model,will
                           cv =6,    # k -cross validation
                            verbose= 2)

## Now we will fit the randomizedSearchCV 
rs_clf.fit(X_train,Y_train)
```

```python
## Checking the Best Model  after doing RandomSearchCV
rs_clf.best_params_
```

### Creating function for Evaluate_preds

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# we will create a function where will pass the Y_test and y_preds as parameter
# and get result of model accuracy, pricision, recall , f1
 
y_preds = rs_clf.predict(X_test)   # this will predict on the X_test and then will compare with actualtrue test value


metric_dic ={}  # creating an empty dict to store all the metric info 
def evaluate_preds(Y_test, y_preds):
    """
    Perforem evalualtion comparsion on Y_true label vs y_preds
    """
    
    accuracy = accuracy_score(Y_test,y_preds)
    precision = precision_score(Y_test,y_preds)
    recall = recall_score(Y_test,y_preds)
    f1 =f1_score(Y_test,y_preds)
    
    metric_dic = {"accuracy" :round(accuracy,2),
                 "precision":round(precision,2),
                 "recall": round(recall,2),
                 "f1":round(f1,2)}
    
    print(f"accuracy:{accuracy*100:.2f}%")
    print(f"precision:{precision*100:.2f}%")
    print(f"recall:{recall*100:.2f}%")
    print(f"f1:{f1*100:.2f}%")
        
    return metric_dic
```

```python
model_metric = evaluate_preds(Y_test,y_preds)
```
